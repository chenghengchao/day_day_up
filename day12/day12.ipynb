- 重复值识别与处理

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data_test04.xlsx')
df.head()
```

- 判断是否存在重复记录：duplicated方法

```python
any(df.duplicated())
```

要有一个结果为True，结果就为True。说明含有重复值。

- 删除重复值：drop_duplicates()

```python
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
any(df.duplicated())
```
