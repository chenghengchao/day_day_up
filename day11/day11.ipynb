- 字符和日期类型的处理

```python
import numpy as np
import pandas as pd
df = pd.read_excel('data_test03.xlsx')
df.head()
```

```python
df.dtypes
```

- 将birthday转换为日期类型:to_datetime方法

```python
df.birthday = pd.to_datetime(df.birthday,format='%Y/%m/%d')
```

```python
df.dtypes
```

- 将手机号转换为字符串：astype方法

```python
df.tel = df.tel.astype('str')
df.dtypes
```

- 新增年龄列：当前日期年份减去生日年份

```python
df['age'] = pd.datetime.today().year - df.birthday.dt.year
df.head()
```

- 新增工龄列：当前日期年份减去开始工作时间年份

```python
df['workage'] = pd.datetime.today().year - df.start_work.dt.year
df.head()
```

- 将手机号中间四位隐藏起来：替换为*

```python
df.tel = df.tel.apply(lambda x: x.replace(x[3:7], '****'))
df.head()
```

- 取出邮箱的域名：split方法

```python
df['email_domain'] = df.email.apply(lambda x:x.split('@')[1])
df.head()
```

- 取出专业信息

```python
df['major'] = df.other.str.findall('专业：(.*?)，')
df.head()
```
