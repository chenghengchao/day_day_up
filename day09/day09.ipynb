5.将Boarding_time转换为时间类型，将New_price转换为浮点类型

```python
sec_cars.Boarding_time = pd.to_datetime(sec_cars.Boarding_time, format='%Y年%m月')
sec_cars.New_price = sec_cars.New_price.str[:-1].astype('float')
sec_cars.dtypes
```

打算使用字符串切片将“万”字剔除，发现报错了

```python
sec_cars.New_price.value_counts()
```

观察发现是因为有141个“暂无”

```python
#自定义一个函数来处理
def convert_price_dtype(x):
    if x=='暂无':
        return 0.0
    else:
        print(x)
        return float(x[:-1])
sec_cars.New_price = sec_cars.New_price.apply(convert_price_dtype)
```
