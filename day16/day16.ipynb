- 数据透视表

pd.pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, dropna=True, margins_name='ALL)

- data:指定数据集
- values：对应“值”
- index：对应“行”
- columns：对应“列”
- aggfunc：指定聚合函数，默认是均值
- fill_value:指定用于填充缺失值的标量
- margins: 是否显示总计，默认是false
- dropna:是否删除缺失值 默认是True
- margins_name:指定总计名称，默认是ALL

```python
import pandas as pd
import numpy as np
```

```python
diamonds = pd.read_table('diamonds.csv', sep=',')
diamonds.head()
```

```python
pd.pivot_table(diamonds, index='color', values='price', margins=True, margins_name='总计')
```

👆透视表，默认取的price的平均值

```python
pd.pivot_table(diamonds, index='clarity', columns='cut', values='carat', aggfunc=np.size, margins=True, margins_name='总计')
```
